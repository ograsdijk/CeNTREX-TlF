```python
from centrex_tlf.hamiltonian import generate_total_reduced_hamiltonian
from centrex_tlf import states, couplings
```

```python
X_states_approx = states.generate_coupled_states_X(
    states.QuantumSelector(J=[0, 1, 2, 3], electronic=states.ElectronicState.X)
)
B_states_approx = states.generate_coupled_states_B(
    states.QuantumSelector(
        J=1, F1=3 / 2, F=2, electronic=states.ElectronicState.B, P=-1
    )
)
```

```python
H_total = generate_total_reduced_hamiltonian(X_states_approx, B_states_approx)
```

```python
df = couplings.generate_br_dataframe(
    H_total.X_states, H_total.B_states, group_ground="J", group_excited=True
)
df
```

```python
df = couplings.generate_br_dataframe(
    H_total.X_states, H_total.B_states, group_ground="J", group_excited=False
)
df
```
